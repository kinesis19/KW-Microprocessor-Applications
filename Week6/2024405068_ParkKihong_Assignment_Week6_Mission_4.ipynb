```python
# Mission 4: LED Control with threading
import RPi.GPIO as GPIO
import time
from threading import Timer

def infinite_LED_Blink():
    GPIO.output(5, True)
    time.sleep(1)
    GPIO.output(5, False)
    Timer(2, infinite_LED_Blink).start()

try:
    GPIO.setmode(GPIO.BCM)
    GPIO.setup(5, GPIO.OUT)
    infinite_LED_Blink()
    
    while True:
        pass
finally:
    print('cleaning up')
    GPIO.cleanup()
```
