```python
# 필요 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt
import cv2

# 폰트 설정
# WARNING:matplotlib.font_manager:findfont: Font family 'Arial' not found. 라고 떠서 해당 내용 주석처리.
# plt.rcParams['font.family'] = 'Arial'
plt.rcParams['axes.unicode_minus'] = False

# 가상 데이터 생성
np.random.seed(0)
# 평균 = 50, 표준편차 = 10의 정규분포 데이터를 1000개 만듦
data = np.random.normal(loc=50, scale=10, size=1000)

# 기초 통계량 계산
mean_value = np.mean(data)
std_value = np.std(data)

print(f"Mean of data: {mean_value:.2f}")
print(f"Standard deviation of data: {std_value:.2f}")

plt.figure(figsize=(10, 5))

# 데이터 전체 분포 확인 (히스토그램으로)
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.title("Data Distribution (Histogram)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

# 데이터 흐름과 패턴 확인 (선 그래프로)
plt.subplot(1, 2, 2)
# 데이터가 1000개라 앞에서부터 100개만 잘라서 그림
plt.plot(data[:100], marker='o', linestyle='-', color='purple', alpha=0.7)
plt.axhline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.title("Sample Data (Line Graph)")
plt.xlabel("Sample Index")
plt.ylabel("Value")
plt.legend()

plt.tight_layout() # 그래프끼리 안 겹치게 여백 조절
plt.show()

# 외부 이미지 불러오기
image_path = "/content/a.png"
image = cv2.imread(image_path)

if image is not None:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.axis('off') # 축(axis) 눈금 지우기
    plt.title("Loaded Image")
    plt.show()
else:
    print("Failed to load the image.")

```
