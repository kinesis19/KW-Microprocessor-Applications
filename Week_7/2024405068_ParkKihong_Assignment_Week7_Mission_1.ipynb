```python
# 센서 정보(강의 센서 정보 + 온도 센서 정보) 1초에 한번 업데이트
import smbus
from imusensor.MPU9250 import MPU9250
from time import sleep

bus = smbus.SMBus(1)
address = 0x68
imu = MPU9250.MPU9250(bus, address)
imu.begin()

print('---------------------------------------------------------------------------------')
while True:
    imu.readSensor()
    imu.computeOrientation()
    print('Ax:{0:<27}  Ay:{1:<27}  Az:{2:<27}'.format(imu.AccelVals[0], imu.AccelVals[1], imu.AccelVals[2]))
    print('Gx:{0:<27}  Gy:{1:<27}  Gz:{2:<27}'.format(imu.GyroVals[0], imu.GyroVals[1], imu.GyroVals[2]))
    print('Mx:{0:<27}  My:{1:<27}  Mz:{2:<27}'.format(imu.MagVals[0], imu.MagVals[1], imu.MagVals[2]))
    print('roll:{0:<25}  pitch:{1:<24}  yaw:{2:<27}'.format(imu.roll, imu.pitch, imu.yaw))
    print('temperature:{0}'.format(imu.Temp))
    print('---------------------------------------------------------------------------------')
    sleep(1)
```
