```python
# Mission5: Interrupt control (without add_event_detect, using gpiozero)
# -> lgpio 사용해야 풀업 스위치 에러 해결됨.
from gpiozero import Button, LED
from signal import pause
import time

led = LED(5)
button = Button(6)

def blink_led():
    print("Button Pressed! Blinking LED")
    led.on()
    time.sleep(0.1)
    led.off()

button.when_pressed = blink_led

print("Ready! Press the button (Ctrl+C to stop)")
try:
    pause()
except KeyboardInterrupt:
    print("cleaning up...")
```
