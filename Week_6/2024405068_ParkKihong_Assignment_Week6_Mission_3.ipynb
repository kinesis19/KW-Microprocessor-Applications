```python
# Mission 3: Control LED Brightness with Keyboard
import RPi.GPIO as GPIO
import time

GPIO.setmode(GPIO.BCM)
GPIO.setup(5, GPIO.OUT)
pwm = GPIO.PWM(5, 100)
pwm.start(50)

try:
    while(True):
        duty_s = input('Brightness (0 to 100): ')
        duty = float(duty_s)  
        pwm.ChangeDutyCycle(duty)
finally:
    print('cleaning up')
    GPIO.cleanup()
```
